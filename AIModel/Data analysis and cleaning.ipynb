# Data analysis and cleaning

#### Importing modules:

```python
import pathlib
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

currentPath = pathlib.Path().resolve()
```

####  Visualization of the BadFaceMask dataset for cleaning decision:

```python
dictSample = { "Nose_Mouth":"", "Mouth_Chin":"", "Chin": "", }

for file in currentPath.glob("BadFaceMask/*/*"):
    if all(dictSample.values()):
        break
    if "Nose_Mouth" in str(file):
        dictSample["Nose_Mouth"] = str(file)
    elif "Mouth_Chin" in str(file):
        dictSample["Mouth_Chin"] = str(file)
    elif "Chin" in str(file):
        dictSample["Chin"] = str(file)

fig, axs = plt.subplots(1, 3, figsize=(15, 15))

for img, ax in zip(dictSample.values(), axs):
    image = Image.open(img)
    image = np.asarray(image)
    ax.imshow(image)
    ax.title.set_text(img.split("\\")[-1])

plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
plt.show()
        
    
```

#### Removing XXXXX_Mask_Nose_Mouth images (Not a realistic scenario):

```python
for file in currentPath.glob('BadFaceMask/*/*'):
    if "Mask_Nose_Mouth" in str(file):
        file.unlink()
```
